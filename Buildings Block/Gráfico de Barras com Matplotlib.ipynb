## Gráfico de Barras com Matplotlib

```python
# Importando a biblioteca
import matplotlib.pyplot as plt

%matplotlib inline
```

### Plotagem na Vertical

```python
figura = plt.figure()
ax = figura.add_axes([0,0,1.5,1])
times = ["São Paulo", "Grêmio", "Santos", "Palmeiras","Cruzeiro",
                  "Internacional","Flamengo","Corinthans","Galo","Vasco"]
titulos = [3, 3, 3, 3, 3, 2, 2, 2, 1, 1]
ax.bar(times, titulos, color = "chocolate") # ax.bar = vertical / ax.barh = horizontal
ax.set_title("Libertadores - Times Brasileiros", fontsize=25)
ax.set_xlabel("Times", fontsize=20)
ax.set_ylabel("Títulos", fontsize=20)
plt.show()
```

### Plotagem na Horizontal

```python
figura = plt.figure()
ax = figura.add_axes([0,0,1.5,1])
times = ["São Paulo", "Grêmio", "Santos", "Palmeiras","Cruzeiro",
                  "Internacional","Flamengo","Corinthans","Atlético Mineiro","Vasco da Gama"]
titulos = [3, 3, 3, 3, 3, 2, 2, 2, 1, 1]
ax.barh(times, titulos, color = "teal") # ax.bar = vertical / ax.barh = horizontal
ax.set_title("Libertadores - Times Brasileiros", fontsize=25)
ax.set_xlabel("Times", fontsize=20)
ax.set_ylabel("Títulos", fontsize=20)
plt.show()
```
