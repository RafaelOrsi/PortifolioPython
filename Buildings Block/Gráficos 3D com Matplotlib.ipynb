## Gráficos 3D com Matplotlib

Para plotar gráficos 3D é necessário importar um submódulo do Matplotlib

```python
from mpl_toolkits import mplot3d

import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Criando ambiente de plotagem do gráfico 3D

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection="3d")

ax.set_title('Ambiente de Plotagem 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

plt.show()
```

### Gráfico de linha no espaço 3D

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

dim_z = np.linspace(0, 18, 1000)
dim_x = np.cos(dim_z)
dim_y = np.sin(dim_z)
ax.plot3D(dim_x, dim_y, dim_z, 'red')
ax.set_title('Espiral de linha no Plano 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

plt.show()
```

### Gráfico de linha e de dispersão no espaço 3D

- Adicionando mais informações na plotagem do gráfico 
- Plotando uma espiral 3d com pontos

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

dim_z = np.linspace(0, 15, 1000)
dim_x = np.cos(dim_z)
dim_y = np.sin(dim_z)
ax.plot3D(dim_x, dim_y, dim_z, 'red')

pontos_z = 15 * np.random.random(120)
pontos_x = np.cos(pontos_z) + 0.1 * np.random.randn(120)
pontos_y = np.sin(pontos_z) + 0.1 * np.random.randn(120)
ax.scatter3D(pontos_x, pontos_y, pontos_z, c=pontos_z, cmap='Spectral')
ax.set_title('Espiral de linha e pontos no Plano 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

plt.show()
```

### Gráficos de contorno no Espaço 3D

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='coolwarm')
ax.set_title('Gráficos de contorno no Espaço 3D', fontsize=18)
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

plt.show()
```

### Gráfico de superfície no Espaço 3D

```python
def f(x, y):
    ar = np.sin(np.sqrt(x ** 2 + y ** 2))
    return ar

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
ax.set_title('Gráfico de superfície no Espaço 3D', fontsize=18);
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

plt.show()
```

### Gráficos de barras no Espaço 3D

```python
import random

# Gráficos de barras 3D
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection="3d")

num_barras = 15
x_pos = random.sample(range(20), num_barras)
y_pos = random.sample(range(20), num_barras)
z_pos = [0] * num_barras
x_size = np.ones(num_barras)
y_size = np.ones(num_barras)
z_size = random.sample(range(20), num_barras)

ax.bar3d(x_pos, y_pos, z_pos, x_size, y_size, z_size, color='cyan')
ax.set_title('Gráfico de barras 3D', fontsize=18);
ax.set_xlabel('Eixo X', fontsize=15)
ax.set_ylabel('Eixo Y', fontsize=15)
ax.set_zlabel('Eixo Z', fontsize=15)

plt.show()
```
